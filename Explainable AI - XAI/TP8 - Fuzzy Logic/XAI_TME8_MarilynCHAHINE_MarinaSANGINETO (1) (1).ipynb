```python
def MPG_crisp(x0, xA, yA, xB, yB, tnorme):

    u = 0
    for i in range(len(xA)-1):
        if xA[i] <= x0 <= xA[i+1]:
            a = (yA[i+1] - yA[i]) / (xA[i+1] - xA[i])
            b = yA[i] - a * xA[i]
            u = a*x0 + b
            break

    Bprime = []
    for v in yB:      # v = μB(xB[i])

        if tnorme == 'LUK':
            Bprime.append(min(1 - u + v, 1))

        elif tnorme == 'REI':
            Bprime.append(1 - u + u*v)

        elif tnorme == 'KLE':
            Bprime.append(max(1 - u, v))

        elif tnorme == 'BRO':
            Bprime.append(1 if u <= v else v)

        elif tnorme == 'GOG':
            Bprime.append(1 if u == 0 else min(v/u, 1))

        elif tnorme == 'WIL':
            Bprime.append(max(1 - u, min(u, v)))

        elif tnorme == 'RES':
            Bprime.append(1 if u <= v else 0)

        elif tnorme == 'MAM':
            Bprime.append(min(u, v))

        elif tnorme == 'LAR':
            Bprime.append(u * v)

    return Bprime
```

```python
def visTnorme_crisp(x0, xA, yA, xB, yB, tnorme):

    # get the range xB to calculate Bprime on
    range_xB = max(xB) - min(xB)
    num_points = max(int(range_xB * 10), 100)
    xB_fine = np.linspace(min(xB) - range_xB/2, max(xB) + range_xB/2, num_points)
    
    # get the values of yB for the added values of xB
    yB_fine = np.interp(xB_fine, xB, yB, left=0, right=0)

    # compute Bprime
    Bprime_fine = MPG_crisp(x0, xA, yA, xB_fine, yB_fine, tnorme)

    plt.figure()
    plt.plot(xB_fine, yB_fine, label="B")
    plt.plot(xB_fine, Bprime_fine, label="B' (MPG_crisp)")
    
    plt.legend()
    plt.xlabel("x in Universe of B")
    plt.ylabel("Membership degree")
    plt.title(f"Modus Ponens Generalisé — {tnorme}")
    plt.grid(True)
    plt.show()
```

```python
def MPG_fuzzy(xA, yA, xB, yB, xA_obs, yA_obs, tnorme):
    
    # xA_obs, yA_obs : fuzzy observation

    Bprime = []

    for i, xb in enumerate(xB):
        # compute xb as sup over all x in xA
        values = []
        for j, xa in enumerate(xA):
            # interpolate fuzzy observation at xa
            u = np.interp(xa, xA_obs, yA_obs, left=0, right=0)  
            mu_A = yA[j]                                       
            mu_B = yB[i]                                      
            
            # compute fuzzy implication and combine with fuzzy observation using t-norm
            if tnorme == 'LUK':
                imp_val = min(1 - mu_A + mu_B, 1)
                val = min(u, imp_val)
            elif tnorme == 'REI':
                imp_val = 1 - mu_A + mu_A * mu_B
                val = 1 - u + u * imp_val
            elif tnorme == 'KLE':
                imp_val = max(1 - mu_A, mu_B)
                val = max(1 - u, imp_val)
            elif tnorme == 'BRO':
                imp_val = mu_B if mu_A > mu_B else 1
                val = 1 if u <= imp_val else imp_val
            elif tnorme == 'GOG':
                imp_val = 1 if mu_A == 0 else min(mu_B / mu_A, 1)
                val = 1 if u == 0 else min(imp_val / u, 1)
            elif tnorme == 'WIL':
                imp_val = max(1 - mu_A, min(mu_A, mu_B))
                val = max(1 - u, min(u, imp_val))
            elif tnorme == 'RES':
                imp_val = 1 if mu_A <= mu_B else 0
                val = 1 if u <= imp_val else 0
            elif tnorme == 'MAM':
                imp_val = min(mu_A, mu_B)
                val = min(u, imp_val)
            elif tnorme == 'LAR':
                imp_val = mu_A * mu_B
                val = u * imp_val
            else:
                print("invalid t-norme")
                
            values.append(val)
        
        # take the supremum over all x
        Bprime.append(max(values))

    return np.array(Bprime)
```

```python
def visTnorme_fuzzy(xA, yA, xB, yB, xA_obs, yA_obs, tnorme):

    # get the range xB to calculate Bprime on
    range_xB = max(xB) - min(xB)
    num_points = max(int(range_xB * 10), 100)
    xB_fine = np.linspace(min(xB) - range_xB/2, max(xB) + range_xB/2, num_points)

    # get the values of yB for the added values of xB
    yB_fine = np.interp(xB_fine, xB, yB, left=0, right=0)

    # compute Bprime
    Bprime_fine = MPG_fuzzy(xA, yA, xB_fine, yB_fine, xA_obs, yA_obs, tnorme)

    plt.figure()
    plt.plot(xB_fine, yB_fine, label="B")
    plt.plot(xB_fine, Bprime_fine, label="B' (MPG_fuzzy)")
    plt.legend()
    plt.xlabel("x in Universe of B")
    plt.ylabel("Membership degree")
    plt.title(f"Generalized Modus Ponens with Fuzzy Observation — {tnorme}")
    plt.grid(True)
    plt.show()
```
